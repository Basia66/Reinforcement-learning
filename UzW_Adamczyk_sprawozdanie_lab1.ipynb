<h3> Zadanie 1 </h3>

<b>Problem uczenia ze wzmocnieniem</b>

Uczenie ze wzmocnieniem polega na stworzeniu agenta (regulator, uczeń), który na podstawie interakcji ze środowiskiem poprawia własną skuteczność (funkcja nagrody -> maksymalizowania nagrody). 
<br>
Do uczenia ze wzmocnieniem nie dostarcza się przygotowanych zestawów trenujących, a jedynie wartościującą informacje trenującą pozyskaną ze środowiska (nagroda).

<b>Relacja między agentem i środowiskiem (RL interaction loop)</b>

Interakcje agenta ze środowiskiem odbywają się w raczej dyskretnych krokach czasu i polegają na obserwowaniu przez agenta kolejnych stanów środowiska oraz wykonywaniu wybranych zgodnie z jego obecną strategią akcji. Po wykonaniu akcji uczeń otrzymuje wartość nagrody/wzmocnienia, które stanowią miarę oceny jakości jego działania.

W każdym kroku czasu:
1. obserwowanie aktualnego stanu
2. wybranie akcji do wykonania w aktualnych stanie
3. wykonanie akcji
4. obserwacja nagrody/wzmocnienia i następnego stanu
5. nauka na podstawie zebranego doświadczenia

<b>Stan środowiska</b>

W uczeniu ze wzmocnieniem środowisko jest niepewne/niekontrolowane i zakłada się, że agent go nie zna. <br>
Środowisko może zmieniać stany oraz dostarczać nagrody pod wpływem wykonywanych przez agenta akcji. <br>
Agent ma wpływ tylko na własne działania,parametry, ale nie ma wpływu na środowisko. <br>

<b>Obserwacja (observation)</b>

Obserwacja ma na celu opisanie obecnego stanu środowiska.

<b>Środowisko częściowo i w pełni obserwowalne (partially vs fully observable)</b>

Kiedy możemy wykryć lub mamy dostęp do każdego stanu agenta w każdym punkcie czasu, mówi się, że jest to w pełni obserwowalne środowisko, w przeciwnym razie jest częściowo obserwowalne.

<b>Przykład:</b> <br>
Szachy – plansza jest w pełni obserwowalna, podobnie jak ruchy przeciwnika. <br>
Jazda samochodem – środowisko jest częściowo obserwowalne, ponieważ to, co jest za rogiem, nie jest znane. <br>

<b>Strategia (policy)</b>

<b>Strategia</b> to funkcja, która przyjmuje obserwację, a zwraca akcję, czyli odpowiada za zachowanie agenta. <br>
Najczęściej stosuje się sieć neuronową. <br>
Agent zawiera dwie strategie: <b>policy</b> i <b>collect policy</b>. <br>

<b>Zadanie epizodyczne a ciągłe (episodic vs continuing task)</b>

<b>Zadania ciągłe</b> to są zadania, które nie mają końca, co oznacza, że nie ma na końcu nagrody. Nagrody są dawane co jakiś czas podczas wykonywania zadania. <br>
Przykład: personal assistance robot
<br><br>
<b>Zadania epizodyczne</b> - mają określony czas działania, gdzie na końcu albo się wygrywa albo przegrywa (przynajmiej jeden końcowy stan). Może być tylko jedna nagroda na końcu zadania, ale można również rozłożyć nagrodę na wszystkie podjęte akcje w zadaniu. <br>
Przykład: car racing video game 
<br><br>
Zadania epizodyczne dzielą się na:
* <b> zadania do-sukcesu </b>
<br>W zadaniach do-sukcesu agent w ramach każdej próby ma do osiągnięcia pewien cel i próba kończy się, kiedy osiągnie on sukces.
* <b> zadania do-porażki </b>
<br>
W zadaniach do-porażki agent stara się unikać pewnej niepożądanej sytuacji. Próba się kończy, kiedy starania te odniosą niepowodzenie.

<b>Funkcja wartości akcji i stanu (action-value function vs state-value function)</b>

<b>Funkcja wartości stanu (state-value funkction)</b> <br>
Funkcja wartości stanu zwraca wartość za wybranie oczekiwanego stanu s (skumulowana nagroda).

gdzie:
$$π - \text{strategia}$$<br>
$$E_π - \text{wartość oczekiwana przy strategii π }$$<br>
$$γ^t- \text{współczynnik dyskontowania(discount factor)}$$<br>
$$r_t- \text{nagroda} $$<br>

<b>Funkcja wartości akcji (action-value funkction)</b> <br>
Funkcja wartości akcji zwraca wartość za użycie akcji a w określonym stanie s (skumulowana nagroda).

gdzie:
$$π - \text{strategia}$$<br>
$$E_π - \text{wartość oczekiwana przy strategii π }$$<br>
$$γ^t- \text{współczynnik dyskontowania(discount factor)}$$<br>
$$r_t- \text{nagroda} $$<br>

<b>Problem eksploracji-eksploatacji (exploration vs exploitation)</b>

<b>Eksploracja</b> (exploration) - działanie w celu pozyskania wiedzy <br>
<b>Eksploatacja</b> (exploitation) - działanie w celu uzyskania nagród <br>

Bardzo trudno jest zastosować eksplorację i eksploatację w tym samym czasie. Można by nawet powiedzieć, że jest to niemożliwe.
